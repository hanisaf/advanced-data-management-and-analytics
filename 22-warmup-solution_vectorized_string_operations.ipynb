1- create a col length that contains the length of text

```python
data['length'] = data['text'].str.len()
data
```

2- create a col words that contains the number of words in text

```python
data['words'] = data['text'].str.split().str.len()
data
```

3- are there any emails in the text

```python
data['text'].str.contains('@').any()
```

4- extract these emails

5- extract tickers and stock markets from text

```python
data[['market', 'ticker']] = data['text'].str.extract(r'\(([A-Z]+):([A-Z]+)\)')
data
```

6- What sectors are more verbose (use more words than others)

```python
#data.groupby('sector').count()
```
