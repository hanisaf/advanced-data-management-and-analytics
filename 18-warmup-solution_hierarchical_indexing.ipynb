This notebook uses the [GapMinder data set](https://www.gapminder.org)

```python
import pandas as pd
data = pd.read_csv("data/gapminder.tsv", sep='\t')
data.head() # show five entries
```

Using multi-indexing, answer the following questions
- Select data about countries in Asia 
- Select data about countries in Asia in years 1952 to 2000
- What is the average `pop` of countries in Asia in years 1952 to 2000

```python
data[data['continent']=='Asia']
```

```python
data.set_index('continent').loc['Asia']#.reset_index()
```

```python
data[(data['continent']=='Asia') & (data['year']>=1952) & (data['year']<=2000)]
```

```python
data.dtypes
```

```python
#data2 = data.set_index(['continent', 'year'])
data2 = data.sort_values(['continent', 'year']).set_index(['continent', 'year'])
data2.head()
```

```python
idx = pd.IndexSlice
```

```python
#data2.sort_index().loc[idx['Asia', 1952:2000], :]
data3 = data2.loc[idx['Asia', 1952:2000], :]#.reset_index()
data3
```

```python
data3['pop'].mean()
```

```python
data2.loc[idx['Asia', 1952:2000], 'pop'].mean()
```
