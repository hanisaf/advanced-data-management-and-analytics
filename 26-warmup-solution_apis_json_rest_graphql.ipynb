# Which repositories were created in 2019?

Use the API endpoint https://api.github.com/repos/:owner/:repo

Replace "owner with username and :repo with the repo

```python
repo_names = [d['full_name'] for d in data]
```

```python
repo_names
```

```python
endpoints = [f"https://api.github.com/repos/{r}" for r in repo_names ]
```

```python
endpoints
```

```python
repo_data = [requests.get(url).json() for url in endpoints]
```

```python
repo_data
```

```python
created = [(rd['name'], rd['created_at']) for rd in repo_data]
```

```python
created
```

```python
df = pd.DataFrame(created)
df.columns = ['repo', 'creation']
df
```

```python
df['date'] = pd.to_datetime(df['creation'])
df
```

```python
df['year'] = df['date'].apply(lambda x: x.year)
df
```

```python
df[df['year']==2019]
```

```python
df.groupby('year')['repo'].count()
```
